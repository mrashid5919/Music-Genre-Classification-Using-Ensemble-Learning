```python
dataset_path = "genres_original/"
csv_path = "data.csv"
SAMPLE_RATE = 22050
DURATION = 30
SAMPLES_PER_TRACK = SAMPLE_RATE * DURATION
```

```python
import os
import csv

def create_genre_csv(dataset_path, csv_path):
    # Open the CSV file for writing
    with open(csv_path, mode="w", newline="") as csv_file:
        csv_writer = csv.writer(csv_file)
        
        # Write the header row
        csv_writer.writerow(["song_name", "genre"])
        
        # Iterate through the genres folder
        for dirpath, dirnames, filenames in os.walk(dataset_path):
            # Ensure we are not at the root level
            if dirpath != dataset_path:
                # Extract genre name (folder name)
                genre_name = os.path.basename(dirpath)
                print(f"Processing genre: {genre_name}")
                
                # Process each audio file in the genre folder
                for file_name in filenames:
                    if file_name.endswith(".wav"):  # Ensure the file is a .wav file
                        # Write the song name and genre to the CSV
                        csv_writer.writerow([file_name, genre_name])
                        print(f"Added: {file_name} -> {genre_name}")
    
    print(f"CSV file created successfully at {csv_path}")
```

```python
dataset_path = "genres_original/"
csv_path = "genre_labels.csv"
create_genre_csv(dataset_path, csv_path)
```
